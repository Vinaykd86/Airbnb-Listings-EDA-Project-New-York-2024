**Data Cleaning**

```python
# Dealing with null values
Data.isnull().sum()
```

```python
# Dropping the null values
Data.dropna(inplace=True)
```

```python
Data.shape
```

```python
# Dealing with duplicates rows
Data.duplicated().sum()
```

```python
Data[Data.duplicated()]
```

```python
# Deleting all duplicated rows
Data.drop_duplicates(inplace=True)
Data.duplicated().sum()
```

```python
# Type casting
Data.dtypes
```

```python
# Changing Data Type
Data["id"]=Data["id"].astype(object)
Data["host_id"]=Data["host_id"].astype(object)

Data.dtypes
```

**Data Analysis**

```python
## Univariate Analysis
Data["price"]
```

```python
# Identifying outliers

df=Data[Data["price"] < 1500]
sns.boxplot( data=df , x="price")
```

```python
# Price Distribution

plt.figure(figsize=(8,5))
sns.histplot(data=df,x="price",bins=100)
plt.title("Price Distribution")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x="availability_365")
plt.title("availability_365_Distribution")
plt.ylabel("frequency")
plt.show()
```

```python
df.groupby(by="neighbourhood_group")["price"].mean()
```

**Feature Engineering**

```python
df["price per bed"]=df["price"]/df["beds"]
df.head()
```

```python
# Average price per bed
df.groupby(by="neighbourhood_group")["price per bed"].mean()
```

```python
## Bi Variate Analysis
```

```python
df.columns
```

```python
sns.barplot(data=df,x="neighbourhood_group", y="price", hue="room_type")
```

```python
# Number of reviews and price relation

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="number_of_reviews",y="price", hue="neighbourhood_group")
plt.title("Locality and Review Analysis")
plt.show()
```

```python
sns.pairplot(data=df, vars=['price','minimum_nights','number_of_reviews'],hue="room_type")
plt.show()
```

```python
## Geographical distribution of airbnb

plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x= "longitude", y="latitude",hue="room_type")
plt.title("Geographical Distribution")
plt.show()
```

```python
##Heat map- correlation of one variable for numerical column

corr=df[["latitude","longitude","price","minimum_nights","number_of_reviews","reviews_per_month","beds"]].corr()
corr

plt.figure(figsize=(8,6))
sns.heatmap(data=corr ,annot=True)
```
